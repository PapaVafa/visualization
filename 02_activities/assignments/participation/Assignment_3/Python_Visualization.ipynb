```python
df = data[data['City'].isin(['Toronto', 'Mississauga', 'Markham', 'North York', 'Richmond Hill'])]
```

```python
sns.countplot(data=df, x='City', palette="Set2")

# Add labels and title
plt.xlabel("City")
plt.ylabel("Count")
plt.title("Number of Schools in different Cities in GTA")

# Show plot
plt.show()
```

```python
sns.histplot(data=df, x='City', hue='School Level', kde=True, binwidth=5, multiple="stack")

# Add title and labels
plt.title('School Levels distribution in GTA')
plt.xlabel('City')
plt.ylabel('Frequency')

# Show plot
plt.show()
```
